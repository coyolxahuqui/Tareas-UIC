```python
import pandas as pd
datos = pd.read_csv("winequalityN.csv")
```

```python
datos = datos.dropna()
#para quitar los que no tienen datos
```

```python
datos.isna().sum()
```

```python
dummies_type = pd.get_dummies(datos["type"], drop_first=True)
```

```python
datos = datos.join(dummies_type)
```

```python
datos.head()
```

```python
datos = datos.drop(["type"], axis=1)
```

```python
datos.head()
```

```python
datos.rename(columns={'white': 'Vino'}, inplace=True)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
X = datos.drop(["Vino"], axis=1)
y = datos["Vino"]
```

```python
from sklearn.model_selection import train_test_split
X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size=.2)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(3, input_shape=(12,), activation='tanh'), #12 características en mi base de datos
    tf.keras.layers.Dense(3, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
print(X_ent.shape)
print(y_ent.shape)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_ent, y_ent, epochs=30, batch_size=1, verbose=1)
```

```python
loss, accuracy = model.evaluate(X_pru, y_pru, verbose=1)
print('Accuracy:', accuracy)
```

```python
import matplotlib.pyplot as plt
```

```python
history = model.fit(X_ent, y_ent, epochs=30, batch_size=1, verbose=1, validation_data=(X_pru, y_pru))
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
