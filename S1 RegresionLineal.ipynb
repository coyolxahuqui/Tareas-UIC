```python
X = df["lsepalo"]
y = df["lpetalo"]
```

```python
type(X)
```

```python
type(X.values)
X.values
```

```python
X.values.reshape(-1,1)
```

```python
X_proc = X.values.reshape(-1,1)
```

```python
y_procesada = y.values.reshape(-1,1)
```

```python
y_procesada
```

```python
from sklearn.model_selection import train_test_split
X_ent, X_pru, y_ent, y_pru = train_test_split(X_proc, y_procesada, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
modelo=LinearRegression()
```

```python
modelo.fit(X_ent, y_ent)
```

```python
import numpy as np
```

```python
modelo.predict(X_pru)
```

```python
predicciones = modelo.predict(X_pru)
```

```python
comparativa = {"PredicciÃ³n": np.squeeze(predicciones), "Valor Real": np.squeeze(y_pru)}
pd.DataFrame(comparativa)
```

```python
print(modelo.score(X_ent, y_ent))
```
